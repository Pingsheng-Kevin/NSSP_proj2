```python
import numpy as np

d = np.load("E:\Datasets\music_data\8275363\sub-001_Resp_Training.npy")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)

d_pca = pca.fit_transform(d.T)

d_pca.shape
```

```python
import matplotlib.pyplot as plt
time_points = np.arange(d_pca.shape[0])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
# ax.scatter(d_pca[:,0], d_pca[:,1], d_pca[:,2], alpha=0.3, c=time_points, cmap='viridis',s=2)
ax.scatter(d_pca[:,1], d_pca[:,2], d_pca[:,3], alpha=0.3, c=time_points, cmap='viridis',s=2)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
